```python
import re
import pandas as pd
```

## Loading Whatsapp Chat in a file

```python
f = open('Chats/WhatsApp Chat with Harshil B.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

## Finding valid Regular Expression 

```python
pattern = re.compile('\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{1,2}\s[A-Z]+[A-Z]\s-\s')
# pattern = re.compile(r"([0-9]+(/[0-9]+)+),\s\d{1,2}:\d{1,2}\s[A-ZA-Z]+\s-\s", re.IGNORECASE)
```

```python
messages = re.split(pattern,data)

messages = messages[1:]
messages
```

```python
dates = re.findall(pattern,data)
# len(dates)
dates
```

```python
df = pd.DataFrame({'user_message':messages,'message_date':dates})

#convert message_date type
df['message_date'] = pd.to_datetime(df['message_date'] , format='%m/%d/%y, %I:%M %p - ')
df.rename(columns={'message_date':'date'} , inplace=True)
df.head()
```

```python
df.shape
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
```

```python
df['user'] = users
df['message'] = messages
df.drop(columns = {'user_message'},inplace=True)
```

```python
df.head()
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
```

```python
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python
df.head()
```

## Analysis for a particular user

```python
selected_user = 'Vishal Verma'
```

```python
def fetch_stats(selected_user , df):
    if(selected_user == 'Overall'):
        # 1. fetch number of messages
        num_messages = df.shape[0]
        #2. fetch number of words
        words = []
        for message in df['message']:
            words.extend(message.split())
        return num_messages,len(words)
    else:
        new_df = df[df['user'] == selected_user]
        num_messages = new_df.shape
        words = []
        for message in new_df['message']:
            words.extend(message.split())
        return num_messages,len(words)
```

```python
num_messages,words = fetch_stats(selected_user,df)
print(num_messages,words)
```

## Counting Most Common Words

```python
# remove group notifications
temp = df[df['user'] != 'group_notification']
```

```python
# remove media omitted
temp = temp[temp['message'] != '<Media omitted>\n']
temp
```

```python
# remove stop words
f = open('stop_hinglish.txt','r')
stop_words = f.read()
```

```python
words = []
for message in temp['message']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)
```

```python
from collections import Counter
pd.DataFrame(Counter(words).most_common(20))
```

## Monthly Timeline

```python
df['num_month'] = df['date'].dt.month
```

```python
timeline = df.groupby(['year','num_month','month']).count()['message'].reset_index()
timeline
```

```python
time = []
for i in range(timeline.shape[0]):
    time.append(timeline['month'][i] + "-" + str(timeline['year'][i]))
```

```python
timeline['time'] = time
timeline
```

```python
plt.plot(timeline['time'],timeline['message'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
df['only_date'] = df['date'].dt.date
```

```python
plt.plot(daily_timeline['only_date'] , daily_timeline['message'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
df['day_name'] = df['date'].dt.day_name()
df['day_name'].value_counts()
```

```python
period = []
for hour in df[['day_name','hour']]['hour']:
    if(hour == 23):
        period.append(str(hour) + "-" + str('00'))
    elif(hour == 0):
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df['time_period'] = period
```

```python
%pip install seaborn
import seaborn as sns
plt.figure(figsize=(20,6))
sns.heatmap(df.pivot_table(index='day_name',columns='time_period',values='message',aggfunc='count').fillna(0))
```
